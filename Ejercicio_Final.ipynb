#Explicación#
Es siguiente codigo se encarga de relver un sudoku de forma determinada

```python
# Una matriz 9x9
sudoku = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

def is_valid(sudoku, row, col, num):
    """Verifica si num se puede colocar en la posición dada."""
    for i in range(9):
        if sudoku[row][i] == num or sudoku[i][col] == num:
            return False
    start_row, start_col = 3 * (row // 3), 3 * (col // 3)
    for i in range(3):
        for j in range(3):
            if sudoku[start_row + i][start_col + j] == num:
                return False
    return True

def find_empty(sudoku):
    """Encuentra una celda vacía (con valor 0) y devuelve su posición."""
    for i in range(9):
        for j in range(9):
            if sudoku[i][j] == 0:
                return i, j
    return None

def print_sudoku(sudoku):
    """Imprime el estado actual del Sudoku."""
    for row in sudoku:
        print(row)
    print()  # Línea en blanco para separar los pasos

def solve_sudoku_greedy(sudoku):
    """Resuelve el Sudoku usando un enfoque voraz."""
    empty = find_empty(sudoku)
    if not empty:
        return True  # Si no hay celdas vacías, el Sudoku está resuelto.
    row, col = empty

    # Generar lista de posibles números y seleccionar el más prometedor
    for num in range(1, 10):
        if is_valid(sudoku, row, col, num):
            sudoku[row][col] = num  # Colocar num en la celda actual
            print_sudoku(sudoku)  # Mostrar el estado después de colocar el número
            if solve_sudoku_greedy(sudoku):
                return True  # Si se resuelve, terminamos
            sudoku[row][col] = 0  # Si no se resuelve, se deshace el cambio
    return False

# Llamada para resolver el Sudoku y mostrar paso a paso
if solve_sudoku_greedy(sudoku):
    print("Sudoku resuelto:")
    print_sudoku(sudoku)  # Imprimir el Sudoku final
else:
    print("No tiene solución")
```
